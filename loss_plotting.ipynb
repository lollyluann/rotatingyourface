```python
file1 = open('yourface-100.txt', 'r')
Lines = file1.readlines()

epochs = []
loss_1 = []
loss_2 = []
losses = []

for line in Lines:
  if line.startswith("Epoch"):
    parts = line.split(", ")
    epoch = parts[0].split(" ")[1]
    loss1 = parts[1].split("=")[1]
    loss2 = parts[2].split("=")[1]
    loss3 = parts[3].split("=")[1]
    epochs.append(int(epoch))
    loss_1.append(eval(loss1))
    loss_2.append(eval(loss2))
    losses.append(eval(loss3))
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (6,5)
l1 = plt.plot(epochs, loss_1)
l2 = plt.plot(epochs, loss_2)
l3 = plt.plot(epochs, losses)
plt.title("Loss over time (100 images)")
plt.legend(["Task 1 Loss", "Task 2 Loss", "Training Loss"], bbox_to_anchor=(1.05, 1))
```
